Became interested in the NIGP code part of the NIGP_DESCRIPTION column. A little research revealed that it's always a 7 digit code in the following format ###-##-##. Futhermore, the code appears to be hierarchical, so that realted items will all have the same first 3 digits.

I will breakout the information in the NIGP_DESCRIPTION into separate columns for anlaysis.

The following function will parse out the NIGP code from the description. I think use the apply function to create a new column with just the NIGP code. Note: The code will be a string.

```python
def NIGP_getcode(df):
    NIGP_d = df["NIGP_DESCRIPTION"]
    while len(NIGP_d) > 7:
        if NIGP_d[7] == ":":
            return NIGP_d[:7]
        else:
            return "9999999"
```

```python
raw_data["NIGP"] = raw_data.apply(NIGP_getcode, axis=1)
```

Having successfully created a new NIGP code-only column, I would like to do the same thing now with the description. The following function separates out the description-only part of NIGP_Description. I think use the apply function to create a new column with just the description.

```python
def NIGP_getdescr(df):
    NIGP_d = df["NIGP_DESCRIPTION"]
    if len(NIGP_d) > 7:
        if NIGP_d[7] == ":":
            dlist = NIGP_d.split(":")[1:]
            dstr = "-".join(dlist)
            return dstr
        else:
            return NIGP_d
    else:
        return NIGP_d
```

Note: The function above uses *dstr = "-".join(dlist)* because I'm splitting on the ":", and the ":" appears one than once in some descriptions. The first time it appears it is associated with the NIGP code and we want to split the line on that. However, the second time it appears it is part of the description and I want to keep both the decription before the ":" and after it. Therefore, I need to rejoin the second, third, etc. parts of the split to make sure I don't lose any of the description.

```python
raw_data["NIGP_DESCR"] = raw_data.apply(NIGP_getdescr, axis=1)
```
